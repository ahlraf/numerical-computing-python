```python
import numpy as np
import pprint as pp
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Gauss-Seidel iteration method

```python
A = np.array([[3., 7., 13.],[1., 5., 3.],[12., 3., -5.]])
B = np.array([76.,28.,1.])
pp.pprint(A)
pp.pprint(B)
```

```python
# Defining our function, which takes 3 arguments: 
# matrix A, solution and matrix B
   
def GaussSeidel(A, x ,B): 
    #Finding length of a(3)        
    n = len(A)                    
    # for loop for 3 times as to calculate x1, x2, x3
    for j in range(0, n):         
        # temp variable d to store B[j] 
        d = B[j]                   
          
        for i in range(0, n):      
            if j != i: 
                d -= A[j,i] * x[i] 
        # updating the value of our solution         
        x[j] = d / A[j,j] 
    # returning our updated solution            
    return x     
```

```python
n = 3                                    
# initial solution depending on n(here n=3)                      
x = np.zeros((3,1))                       

maxiter = 100

#loop run for m times depending on m the error value 
gauss_table1 = []
for i in range(1, maxiter+1):             
    x = GaussSeidel(A, x, B)
    #print each time the updated solution 
    print("Iteration #",i)
    gauss_table1.append((i,x[0][0]))
    print(x)
    print('\n')
```

Solution does not converge, the matrix may not be diagonally dominant

**Test for whether A is a diagonally dominant matrix or not**

```python
# function to check whether the given   
# matrix is diagonally dominant or not. 
def isDDM(m, n) : 
  
    # for each row 
    for i in range(0, n) :          
      
        # for each column, finding 
        # sum of each row. 
        sum = 0
        for j in range(0, n) : 
            sum += abs(m[i,j])      
  
        # removing the  
        # diagonal element. 
        sum -= abs(m[i,i]) 
  
        # checking if diagonal  
        # element is less than  
        # sum of non-diagonal 
        # element. 
        if (abs(m[i,i]) < sum) : 
            return False
  
    return True
```

```python
isDDM(A,n)
```

We now use row permutations to convert A to a diagonally dominant matrix.



```python
import random
isdom = isDDM(A,n)
while 1:
    if isDDM(A,n):
        print('Matrix A is diagonally dominant')
        break
    else:
        A = np.random.permutation(A)
```

Checking number of iterations needed for convergence

```python
def GSiterations(A, b, x, N, tol=1E-9):
    maxIterations = 1000000
    xprev = [0.0 for i in range(N)]
    for i in range(maxIterations):
        for j in range(N):
            xprev[j] = x[j]
        for j in range(N):
            summ = 0.0
            for k in range(N):
                if (k != j):
                    summ = summ + A[j][k] * x[k]
            x[j] = (b[j] - summ) / A[j][j]
        diff1norm = 0.0
        oldnorm = 0.0
        for j in range(N):
            diff1norm = diff1norm + abs(x[j] - xprev[j])
            oldnorm = oldnorm + abs(xprev[j])  
        if oldnorm == 0.0:
            oldnorm = 1.0
        norm = diff1norm / oldnorm
        if (norm < tol) and i != 0:
            print("Sequence converges to [", end="")
            for j in range(N - 1):
                print(round(x[j]), ",", end="")
            print(round(x[N - 1]), "]. Took", i + 1, "iterations.")
            return
    print("Doesn't converge.")

x0 = [0.0,0.0,0.0]
GSiterations(A,B,x0,3)
```

### Jacobi iteration method

```python
A = np.array([[3., 7., 13.],[1., 5., 3.],[12., 3., -5.]])
B = np.array([76.,28.,1.])
pp.pprint(A)
pp.pprint(B)
```

```python
from numpy.linalg import *

# Defining our function
def jacobi(A, b, x0, tol, jacobi_table, maxiter=200):
    """
    Performs Jacobi iterations to solve the line system of
    equations, Ax=b, starting from an initial guess, "x0".

    Terminates when the change in x is less than "tol", or
    if "maxiter" [default=200] iterations have been exceeded.

    Returns 3 variables:
        1.  x, the estimated solution
        2.  rel_diff, the relative difference between last 2
            iterations for x
        3.  k, the number of iterations used.  If k=maxiter,
            then the required tolerance was not met.
    """
    n = A.shape[0]
    x = x0.copy()
    x_prev = x0.copy()
    k = 0
    rel_diff = tol * 2

    while (rel_diff > tol) and (k < maxiter):

        for i in range(0, n):
            subs = 0.0
            for j in range(0, n):
                if i != j:
                    subs += A[i,j] * x_prev[j]

            x[i] = (b[i] - subs ) / A[i,i]
        k += 1

        rel_diff = norm(x - x_prev) / norm(x)
        print('Iteration #',k)
        pp.pprint(x)
        jacobi_table.append((k,x[0][0]))
        print('Relative difference =',rel_diff,'\n')
        x_prev = x.copy()

    return x, rel_diff, k, jacobi_table

# Main code starts here
# ---------------------
GL = 1.6
d = 0.8

x0 = np.zeros((3,1));

tol = 1E-9
maxiter = 200
jacobi_table1 = []
x, rel_diff, k, jacobi_table1 = jacobi(A, B, x0, tol, jacobi_table1, maxiter)
if k == maxiter:
    print(('WARNING: the Jacobi iterations did not '
           'converge within the required tolerance.'))
print(('The solution is \n%s; \nwithin a tolerance of %g, '
        'using %d iterations.' % (x, rel_diff, k)))
print('Solution error = norm(Ax-b) = %g' % \
            norm(np.dot(A,x)-B)) 
print('Condition number of A = %0.5f' % cond(A))
print('Solution from built-in functions = %s' % solve(A, B))
```

Solution does not converge, because as seen before, A is not diagonally dominant.
